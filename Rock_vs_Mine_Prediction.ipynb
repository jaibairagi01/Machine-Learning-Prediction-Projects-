**Supervised Learning:**
Logistic Regression Model

```python
#import the dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split #to train and test the model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score #to find the accuracy of model
```

Data Collection & Data Processing


```python
#loading the dataset
sonar_data = pd.read_csv('/content/sonar data.csv', header = None)
```

```python
sonar_data.head()
```

```python
#Number of rows and columns
sonar_data.shape
```

```python
sonar_data.describe() #describe --> statistical measures of the data
```

```python
sonar_data[60].value_counts()
```

M --> Mine
R --> Rock

```python
sonar_data.groupby(60).mean()
```

```python
#Seprating data and Labels
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]
```

```python
print(X)
print(Y)
```

Training and Testing Data

```python
#Splitting a data into training and testing
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify = Y, random_state = 1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

Model Training --> Logistic Regression Model

```python
model = LogisticRegression()
```

```python
#training the Logistic Regression Model with training data
model.fit(X_train, Y_train)
```

Model Evaluation


```python
#accuracy on the training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy Score on Training Data' , training_data_accuracy)
```

```python
#accuracy on the test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy Score on Test Data' , test_data_accuracy)
```
